```python
"""
Author: CHANDRAMOHAN T N
File: Many Models.py 
"""

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import precision_recall_fscore_support
from sklearn import metrics
from threading import Thread
from Queue import Queue

def Get_data(i_f):
    f = open(i_f, 'r')
    data = []
    while 1:
        line = f.readline()
        line = line[0:-1]
        if len(line) > 0:
            items = map(int, line.split(','))
            data.append(items)
        else:
            f.close()
            break
    return data

def Get_labels(i_f):
    f = open(i_f, 'r')
    labels = []
    while 1:
        line = f.readline()
        line = line[0:-1]
        if len(line) > 0:
            labels.append(int(line))
        else:
            f.close
            break
    return labels

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def log_likelihood(x, y, w):
    p = sigmoid(np.dot(x, w))
    loss = -y * np.log(p) - (1 - y) * np.log(1 - p)
    return loss.mean()

def gradient(x, y, w):
    p = sigmoid(np.dot(x, w))
    grad = np.dot(np.transpose(x), (p - y))
    return grad.reshape(w.shape)

def plot(a, b, c, x, l):
    plt.plot(x, a, 'r', label='AUC Loss')
    plt.plot(x, b, 'g', label='Log Loss')
    plt.plot(x, c, 'b', label='Sq Error')
    plt.title(l)
    plt.legend(loc='upper right')
    plt.ylabel('Loss')
    plt.xlabel('Iteration')
    plt.savefig('Many Models Test Loss.png')
    plt.show()
    plt.close()

class Log_reg(object):

    def __init__(self, n):
        self.w = np.zeros((n, 1))

    def train_ogd(self, l_rate, x, y):
        self.w = self.w - l_rate * gradient(x, y, self.w)
        return self.w

    def update_weights(self, w):
        self.w = w

    def save_model(self):
        np.savez_compressed('Model_many_models', w=self.w)

    def predict(self, x):
        labels = [0 for i in xrange(len(x))]
        p = sigmoid(np.dot(x, self.w))
        for i in xrange(len(x)):
            if p[i] > 0.5:
                labels[i] = 1
        return p, labels

    def evaluate(self, y_pred, y_true, test_x):
        fpr, tpr, thresholds = metrics.roc_curve(y_true, y_pred)
        auc = metrics.auc(fpr, tpr)
        auc_loss = 1 - auc
        log_loss = log_likelihood(test_x, y_true, self.w)
        sq_error = np.sum((y_pred - y_true)**2) / y_pred.shape[0]
        return auc_loss, log_loss, sq_error
    
    def perf_metrics(self, y_real, y_pred, y_prob):
        precision, recall, thresholds = precision_recall_curve(y_real, y_prob)
        plt.plot(precision, recall, 'r')
        plt.ylabel('Precision')
        plt.xlabel('Recall')
        plt.savefig('PR_curve_many_models.png')
        plt.show()
        print('Accuracy:')
        print(accuracy_score(y_real, y_pred))
        print('Precison : Recall : F-score')
        print(precision_recall_fscore_support(y_real, y_pred))

def main():
    i_f = '../Data/books/Train_x.csv'
    train_x = Get_data(i_f)
    train_x = np.concatenate((np.ones((len(train_x), 1)), np.array(train_x)), axis=1)
    i_f = '../Data/books/Test_x.csv'
    test_x = Get_data(i_f)
    test_x = np.concatenate((np.ones((len(test_x), 1)), np.array(test_x)), axis=1)
    i_f = '../Data/books/Train_y.csv'
    train_y = Get_labels(i_f)
    train_y = np.reshape(train_y, (train_x.shape[0], 1))
    i_f = '../Data/books/Test_y.csv'
    test_y = Get_labels(i_f)
    test_y = np.reshape(test_y, (test_x.shape[0], 1))

    l_reg1 = Log_reg(train_x.shape[1])
    l_reg2 = Log_reg(train_x.shape[1])
    l_reg3 = Log_reg(train_x.shape[1])
    l_reg4 = Log_reg(train_x.shape[1])
    l_reg5 = Log_reg(train_x.shape[1])
    l_reg6 = Log_reg(train_x.shape[1])
    l_reg7 = Log_reg(train_x.shape[1])
    l_reg8 = Log_reg(train_x.shape[1])
    l_reg9 = Log_reg(train_x.shape[1])
    l_reg10 = Log_reg(train_x.shape[1])
    max_iter = 15000
    te_auc_loss = []
    te_log_loss = []
    te_sq_error = []
    for i in range(max_iter):
        idx = np.random.randint(len(train_x))
        q1, q2, q3, q4, q5, q6, q7, q8, q9, q10 = Queue(), Queue(), Queue(), Queue(), Queue(), Queue(), Queue()
        t1 = Thread(target=lambda q1, arg1, arg2: q1.put(l_reg1.train_ogd(arg1, arg2)), args=(q1, 0.01, idx))
        t2 = Thread(target=lambda q2, arg1, arg2: q2.put(l_reg2.train_ogd(arg1, arg2)), args=(q2, 0.02, idx))
        t3 = Thread(target=lambda q3, arg1, arg2: q3.put(l_reg3.train_ogd(arg1, arg2)), args=(q3, 0.03, idx))
        t4 = Thread(target=lambda q4, arg1, arg2: q4.put(l_reg4.train_ogd(arg1, arg2)), args=(q4, 0.04, idx))
        t5 = Thread(target=lambda q5, arg1, arg2: q5.put(l_reg5.train_ogd(arg1, arg2)), args=(q5, 0.05, idx))
        t6 = Thread(target=lambda q6, arg1, arg2: q6.put(l_reg6.train_ogd(arg1, arg2)), args=(q6, 0.06, idx))
        t7 = Thread(target=lambda q7, arg1, arg2: q7.put(l_reg7.train_ogd(arg1, arg2)), args=(q7, 0.07, idx))
        t8 = Thread(target=lambda q8, arg1, arg2: q8.put(l_reg8.train_ogd(arg1, arg2)), args=(q8, 0.08, idx))
        t9 = Thread(target=lambda q9, arg1, arg2: q9.put(l_reg9.train_ogd(arg1, arg2)), args=(q9, 0.09, idx))
        t10 = Thread(target=lambda q10, arg1, arg2: q10.put(l_reg10.train_ogd(arg1, arg2)), args=(q10, 0.1, idx))
        t1.start()
        t2.start()
        t3.start()
        t4.start()
        t5.start()
        t6.start()
        t7.start()
        t8.start()
        t9.start()
        t10.start()
        l_reg1.update_weights(q1.get())
        l_reg2.update_weights(q2.get())
        l_reg3.update_weights(q3.get())
        l_reg4.update_weights(q4.get())
        l_reg5.update_weights(q5.get())
        l_reg6.update_weights(q6.get())
        l_reg7.update_weights(q7.get())
        l_reg8.update_weights(q8.get())
        l_reg9.update_weights(q9.get())
        l_reg10.update_weights(q10.get())
        if i % 500 == 0:
            print('Iteration: ' + str(i))
    t1.join()
    t2.join()
    t3.join()
    t4.join()
    t5.join()
    t6.join()
    t7.join()
    t8.join()
    t9.join()
    t10.join()
    np.savez_compressed('Model_many_1', w=l_reg1.w)
    np.savez_compressed('Model_many_2', w=l_reg2.w)
    np.savez_compressed('Model_many_3', w=l_reg3.w)


if __name__ == '__main__':
    main()
```
